```python
from statistics import mean
imgname1 = '57노7660_370_a37159f6-8fe5-11ed-8164-ac74b1bf3c26_ruby'
index = 0
clahe = cv2.createCLAHE()
samemean = []
diffmean = []
for i in jpg_filenames:    
    image2 = cv2.imread(root_path + '\\'+imgname1+'.jpg')
    image1 = cv2.imread(root_path + '\\'+i+'.jpg')
    """gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
    image1 = clahe.apply(gray1)
    image2 = clahe.apply(gray2)"""

    # Convert the images to grayscale
    #gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    #gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # Create a SIFT object
    sift = cv2.xfeatures2d.SIFT_create()

    # Detect the keypoints and compute the descriptors
    keypoints1, descriptors1 = sift.detectAndCompute(image1, None)
    keypoints2, descriptors2 = sift.detectAndCompute(image2, None)
    

    # Draw the top 50 matches
    #image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

    # Show the image with the matches
    #cv2.imshow("Matches", image_matches)
   
    # Set up the FLANN parameters
    FLANN_INDEX_KDTREE = 1
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks = 50)

    # Create a FLANN matcher object
    flann = cv2.FlannBasedMatcher(index_params, search_params)

    # Match the descriptors
    matches = flann.knnMatch(descriptors1, descriptors2, k=2)
    num_matches = len(matches)

    # Filter out the good matches
    good_matches = []

    # ratio test as per Lowe's paper
    for k, (m, n) in enumerate(matches):
        if m.distance < 0.8 * n.distance:
            good_matches.append(m)

    # Get the number of good matches
    num_good_matches = len(good_matches)

    # Draw the top 50 matches
    image_matches = cv2.drawMatches(image1, keypoints1, image2, keypoints2, good_matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    cv2.imwrite(save_path + '\\'+f"{imgname1}&{i}_matches.jpg", image_matches)

    if imgname1.split('_')[0] == i.split('_')[0]:
        samemean.append(len(good_matches)/len(matches))
        print("SameImg============:: Number of matches:", num_matches)
        print({index}, "image::", imgname1.split('_')[0], "and", i.split('_')[0],"Number of good matches:", num_good_matches, '비율::', len(good_matches)/len(matches))
        with open(save_path + '\\'+f"{index}_{imgname1.split('_')[0]}_{i.split('_')[0]}&{i}_matches.txt", "w") as f:
            f.write("imagename1: {}\n".format(imgname1))
            f.write("imagename2: {}\n".format(i))
            # Write the number of matches and good matches to the file
            f.write("Number of matches: {}\n".format(num_matches))
            f.write("Number of good matches: {}\n".format(num_good_matches))
    else:
        diffmean.append(len(good_matches)/len(matches))
        print("Number of matches:", num_matches)
        print({index}, "image::", imgname1.split('_')[0], "and", i.split('_')[0],"Number of good matches:", num_good_matches, '비율::', len(good_matches)/len(matches))
        with open(save_path + '\\'+f"{index}_{imgname1.split('_')[0]}_{i.split('_')[0]}&{i}_matches.txt", "w") as f:
            # Write the number of matches and good matches to the file
            f.write("imagename1: {}\n".format(imgname1))
            f.write("imagename2: {}\n".format(i))
            f.write("Number of matches: {}\n".format(num_matches))
            f.write("Number of good matches: {}\n".format(num_good_matches))
        
        
    index = index+1
    
print("samemean::", mean(samemean))
print("diffmean::", mean(diffmean))
```

```python
import cv2
import numpy as np

# Load the two images
img1 = cv2.imread("image1.jpg")
img2 = cv2.imread("image2.jpg")

# Convert the images to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Initialize the ORB detector
orb = cv2.ORB_create()

# Find keypoints and descriptors with ORB
keypoints1, descriptors1 = orb.detectAndCompute(gray1, None)
keypoints2, descriptors2 = orb.detectAndCompute(gray2, None)

# Create FLANN matcher
FLANN_INDEX_LSH = 6
index_params = dict(algorithm=FLANN_INDEX_LSH, table_number=6, key_size=12, multi_probe_level=1)
search_params = dict(checks=50)
matcher = cv2.FlannBasedMatcher(index_params, search_params)

# Match descriptors
matches = matcher.match(descriptors1, descriptors2)

# Sort the matches in order of distance
matches.sort(key=lambda x: x.distance)

# Draw the top N matches
N = 10
image = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches[:N], img2.copy(), flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Show the image
cv2.imshow("Matching points", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
